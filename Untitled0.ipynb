```python
import os
from flask import Flask, render_template, send_from_directory # Import send_from_directory
from fund_performance_app.data_processor import load_and_clean_data, analyze_performance_by_type
from flask_ngrok import run_with_ngrok
import matplotlib.pyplot as plt # Import matplotlib
import seaborn as sns # Import seaborn
import pandas as pd # Import pandas (already imported, but good to be explicit)

# Define the Flask application instance
# Specify template folder relative to the app.py location within fund_performance_app
app = Flask(__name__, template_folder='templates', static_folder='static') # Add static folder

# Wrap the Flask app with run_with_ngrok to expose it via ngrok
run_with_ngrok(app)

@app.route('/')
def index():
    funds_df = load_and_clean_data()

    performance_html = ""
    chart_files = [] # List to store chart image filenames

    if funds_df.empty:
        performance_html = "<p>Error loading or processing data.</p>"
    else:
        performance_results = analyze_performance_by_type(funds_df)

        if performance_results.empty:
            performance_html = "<p>No performance data available after analysis.</p>"
        else:
            # Prepare data to pass to the template (convert DataFrame to HTML table)
            performance_html = performance_results.to_html(classes='table table-striped')

            # Generate and save charts
            performance_cols = ['1 Ay (%)', '3 Ay (%)', '6 Ay (%)', 'Yılbaşı (%)', '1 Yıl (%)', '3 Yıl (%)', '5 Yıl (%)']
            sns.set_style('whitegrid')

            # Ensure static directory exists to save charts
            static_dir = os.path.join(app.root_path, 'static')
            if not os.path.exists(static_dir):
                os.makedirs(static_dir)

            for col in performance_cols:
                if col in performance_results.columns:
                    plt.figure(figsize=(12, 6))
                    sns.barplot(x=performance_results.index, y=performance_results[col])
                    plt.title(f'Ortalama {col} Performansı (Fon Türüne Göre)')
                    plt.xlabel('Şemsiye Fon Türü')
                    plt.ylabel(f'Ortalama {col} (%)')
                    plt.xticks(rotation=45, ha='right')
                    plt.tight_layout()

                    # Save the plot to the static directory
                    chart_filename = f'performance_by_type_{col.replace(" (%)", "").replace(" ", "_")}.png'
                    chart_path = os.path.join(static_dir, chart_filename)
                    plt.savefig(chart_path)
                    plt.close() # Close the plot figure

                    chart_files.append(chart_filename) # Add filename to the list

    # Render the index.html template, passing the processed data and chart filenames
    return render_template('index.html', performance_html=performance_html, chart_files=chart_files)

# Add a route to serve static files (charts) - Flask handles this automatically with static_folder, but being explicit is good.
# @app.route('/static/<path:filename>')
# def static_files(filename):
#     return send_from_directory('static', filename)


if __name__ == '__main__':
    # This will now run the app with ngrok
    app.run()
```
